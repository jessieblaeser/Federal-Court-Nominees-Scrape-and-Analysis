```python
import requests
from bs4 import BeautifulSoup

#here, you're doing you regular BeautifulSoup commands, but for a range of numbers
    #these numbers represent page numbers on the Senate Judiciary page 
nom_links = []
for x in range(1,27): 
    my_url = 'https://www.judiciary.senate.gov/hearings?PageNum_rs=' + str(x)
    raw_html = requests.get(my_url).content
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    link_info = soup_doc.find_all(class_='url summary pull-left')
    for info in link_info: 
        if "Nominations" in info.text.strip():
            nom_links.append(info['href'])

len(nom_links)
```

```python
import re
```

```python
nominee_info = []
for link in nom_links: 
    my_url = 'https://www.judiciary.senate.gov/' + link
    raw_html = requests.get(my_url).content
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    hearing_pdf = soup_doc.find_all(class_='hearing-pdf')  
    for info in hearing_pdf:
        nominee_dict = {}
        if "sjq" in info['href']:
#             print(info.get_text())
#             print(info['href'])
#             print('---------')
            nominee_dict['pdf_name'] = info.get_text()
            my_link = info['href'].split('/')
            nominee_dict['sjq_link'] = 'https://www.judiciary.senate.gov/' + my_link[-1]
            nominee_info.append(nominee_dict)
        else: 
            pass
#nominee_info
```
