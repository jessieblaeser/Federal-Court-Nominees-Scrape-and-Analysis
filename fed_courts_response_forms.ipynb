```python
import requests
from bs4 import BeautifulSoup
nom_links = []
for x in range(1,10): # there are actually 81
    my_url = "https://www.judiciary.senate.gov/hearings?PageNum_rs=" + str(x)
    raw_html = requests.get(my_url).content
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    each_row = soup_doc.find_all(class_='url summary pull-left')
    for row in each_row: 
        if "Nominations" in row.string.strip(): 
            nom_links.append(row["href"])
len(nom_links)
#nom_links = all nomination hears that have taken place during biden's presidency (so far) 
    # and trump's presidency overall would be range(1,24) i think)
    
    #40
```

```python
import re
```

```python

link_info = []
for urls in nom_links: 
    my_url = "https://www.judiciary.senate.gov/" + urls
    raw_html = requests.get(my_url).content
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    hearing_links = soup_doc.find_all(class_='hearing-pdf')
    for name in hearing_links: 
        link_dict = {}
        link_dict["name"] = name.get_text()
        link_dict["pdf_link"] = name["href"]
        if (re.search("Responses", link_dict["name"])):
            link_info.append(link_dict)
            
link_info
#166
#171
```
